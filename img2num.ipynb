```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("/content/cat_gray-1.jpg")
plt.imshow(img)
```

```python
# 画像の高さと幅を取得
width, height  = img.size
print(width, height)
```

```python
# 画像の特定の場所のピクセルの色を数値として取得
img.getpixel((99, 99))
```

```python
#画像サイズと同じ大きさの配列を作成
image_array = np.empty((height, width), dtype=int)
image_array
```

```python
# 画像を数値に変換
for y in range(height):
    for x in range(width):
        # ピクセルの色の数値を代入している
        image_array[y][x] = img.getpixel((x, y))

image_array
```

```python
plt.imshow(image_array)
```
